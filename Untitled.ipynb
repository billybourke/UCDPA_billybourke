```python
# SECTION 1

#Importing defensive statistical from CSV files.
#Merging, concatenating, reindexing, replacing missing values.
#Numpy function to address change in team names (SDG/LAC, STL/LAR, OAK/LVR)

import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None
pd.set_option("display.max_rows", None, "display.max_columns", None)

url = 'https://www.pro-football-reference.com/years/2021/passing.htm'
qb21 = pd.read_html(url)[0]
qb21 = qb21.drop(labels=[29, 60, 91])
qb21['Year'] = 2021

url = 'https://www.pro-football-reference.com/years/2020/passing.htm'
qb20 = pd.read_html(url)[0]
qb20 = qb20.drop(labels=[29, 60, 91])
qb20['Year'] = 2020

url = 'https://www.pro-football-reference.com/years/2019/passing.htm'
qb19 = pd.read_html(url)[0]
qb19 = qb19.drop(labels=[29, 60, 91])
qb19['Year'] = 2019

url = 'https://www.pro-football-reference.com/years/2018/passing.htm'
qb18 = pd.read_html(url)[0]
qb18 = qb18.drop(labels=[29, 60, 91])
qb18['Year'] = 2018

url = 'https://www.pro-football-reference.com/years/2017/passing.htm'
qb17 = pd.read_html(url)[0]
qb17 = qb17.drop(labels=[29, 60, 91])
qb17['Year'] = 2017

url = 'https://www.pro-football-reference.com/years/2016/passing.htm'
qb16 = pd.read_html(url)[0]
qb16 = qb16.drop(labels=[29, 60, 91])
qb16['Year'] = 2016

url = 'https://www.pro-football-reference.com/years/2015/passing.htm'
qb15 = pd.read_html(url)[0]
qb15 = qb15.drop(labels=[29, 60])
qb15['Year'] = 2015

url = 'https://www.pro-football-reference.com/years/2014/passing.htm'
qb14 = pd.read_html(url)[0]
qb14 = qb14.drop(labels=[29, 60, 91])
qb14['Year'] = 2014

url = 'https://www.pro-football-reference.com/years/2013/passing.htm'
qb13 = pd.read_html(url)[0]
qb13 = qb13.drop(labels=[29, 60])
qb13['Year'] = 2013

url = 'https://www.pro-football-reference.com/years/2012/passing.htm'
qb12 = pd.read_html(url)[0]
qb12 = qb12.drop(labels=[29, 60])
qb12['Year'] = 2012


qb_total = pd.concat([qb21, qb20, qb19, qb18, qb17, qb16, qb15, qb14, qb13, qb12], ignore_index=False)
qb_total['Team'] = qb_total['Tm']
qb_total['Tm'] = np.where(qb_total['Tm'] == 'OAK', 'LVR', qb_total['Tm'])
qb_total['Tm'] = np.where(qb_total['Tm'] == 'SDG', 'LAC', qb_total['Tm'])
qb_total['Tm'] = np.where(qb_total['Tm'] == 'STL', 'LAR', qb_total['Tm'])
qb_total = qb_total.reindex(columns=['Player', 'Tm', 'Year', 'Age', 'G', 'GS', 'QBrec', 'Cmp', 'Att',
       'Cmp%', 'Yds', 'TD', 'TD%', 'Int', 'Int%', '1D', 'Lng', 'Y/A', 'AY/A',
       'Y/C', 'Y/G', 'Rate', 'QBR', 'Sk', 'Yds.1', 'Sk%', 'NY/A', 'ANY/A',
       '4QC', 'GWD', 'Pos', 'Rk'])
qb_total['Player'] = qb_total['Player'].str.replace('\W', '', regex=True)
qb_total['4QC'] = qb_total['4QC'].fillna('0')
qb_total['GWD'] = qb_total['GWD'].fillna('0')
qb_total[['QBwin', 'QBloss', 'QBdraw']] = qb_total['QBrec'].str.split('-', expand=True)
qb_total[['Rk', 'Age', 'G', 'GS', 'Cmp', 'Att', 'Cmp%', 'Yds', 'TD', 'TD%', 'Int', 'Int%', '1D', 'Lng', 'Y/A', 'AY/A', 'Y/C',
          'Y/G', 'Rate', 'QBR', 'Sk', 'Yds.1', 'Sk%', 'NY/A', 'ANY/A', '4QC', 'GWD', 'QBwin', 'QBloss', 'QBdraw']] = qb_total[['Rk',
          'Age', 'G', 'GS', 'Cmp', 'Att', 'Cmp%', 'Yds', 'TD', 'TD%', 'Int', 'Int%', '1D', 'Lng', 'Y/A', 'AY/A',
          'Y/C', 'Y/G', 'Rate', 'QBR', 'Sk', 'Yds.1', 'Sk%', 'NY/A', 'ANY/A', '4QC', 'GWD', 'QBwin', 'QBloss',
          'QBdraw']].apply(pd.to_numeric)
qb_total['W/L%'] = qb_total['QBwin'] / qb_total['GS']
qb_total['W/L%'] = qb_total['W/L%'].round(3)
qb_total = qb_total.sort_values(by=['W/L%'], ascending=False)
qb_total = qb_total[qb_total['Pos'] == 'QB']
qb_total = qb_total[qb_total['GS'] >= 1]
qb_total = qb_total.set_index('W/L%', drop=False)
```

```python
# SECTION 10: Replace 'QBR' with any of the below statistics.

# Higher = better: Cmp', 'Cmp%', 'Yds', 'TD', 'TD%',  '1D', 'Lng', 'Y/A', 'AY/A', 'Y/C', 'Y/G', 'Rate', 'QBR',
#                 'NY/A', 'ANY/A', '4QC', 'GWD', 'W/L%'
# Lower = better (replace 'nlargest' with 'nsmallest', line 9): 'Int', 'Int%', Sk, Sk%
# Below shows teams - for individual QBs, replace 'Tm' with 'Player', line 10.

qb_groups = qb_total.reset_index(drop=True)
def quarterback(stat):
    global qb_groups
    qb_groups = qb_groups.nlargest(150,[stat])
    qb_groups = qb_groups.groupby('Tm').size().sort_values(ascending=False).reset_index(name='count')
    return(qb_groups)

print(quarterback('QBR'))
```

```python
# SECTION 11: Replace 'PA' with any of the below statistics.

# Higher = better (replace 'nsmallest' with 'nlargest'): TO, FL, Int, TO%, W, W-L%
# Lower = better: 'PA', 'Yds', 'Ply', 'Y/P', '1stD', 'PassCmp', 'PassYds', 'PassTD', 'NY/A', 'RunYds', 'RunTD',
#                 'Y/A', 'Pen', 'PenYds', 'Sc%', 'L'

defense_groups = defense_total.reset_index(drop=True)
def defense(stat):
    global defense_groups
    defense_groups = defense_groups.nsmallest(150,[stat])
    defense_groups = defense_groups.groupby('Tm').size().sort_values(ascending=False).reset_index(name='count')
    return(defense_groups)

print(defense('PA'))
```
